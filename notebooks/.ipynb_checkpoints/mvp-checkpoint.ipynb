```python
import numpy as np
from scipy import signal
import matplotlib.pyplot as plt
import math as mt
```

```python
in_array = np.arange(17.*50)
radian_Values = np.radians(in_array) 
```

```python
sig = np.cos(radian_Values)
```

```python
win = signal.hann(50)
filtered = signal.convolve(sig, win, mode='same') / sum(win)
```

```python
fig, (ax_orig, ax_win, ax_filt) = plt.subplots(3, 1, sharex=True)
ax_orig.plot(sig)
ax_orig.set_title('Original pulse')
ax_orig.margins(0, 0.1)
ax_win.plot(win)
ax_win.set_title('Filter impulse response')
ax_win.margins(0, 0.1)
ax_filt.plot(filtered)
ax_filt.set_title('Filtered signal')
ax_filt.margins(0, 0.1)
fig.tight_layout()
fig.show()
```

```python
fig, ax_orig = plt.subplots(1, 1, sharex=True)
ax_orig.plot(sig)
ax_orig.plot(filtered)
ax_orig.set_title('Original pulse')
```

```python
from scipy.fftpack import fft
# Number of sample points
N = 5000
# sample spacing
T = 1.0 / 800.0

# Parameters
#  frequency
f1 = 50
f2 = 100
f3 = 140
f4 = 250

w1 = 2.0*np.pi*f1
w2 = 2.0*np.pi*f2
w3 = 2.0*np.pi*f3
w4 = 2.0*np.pi*f4

#  phase
th1 = 0
th2 = 0
th3 = 0
th4 = 0

#  amplitude
A1 = 1
A2 = 1
A3 = 0.5
A4 = 0.2

# Weight construction
s1 = A1*np.cos(w1*x + th1)
s2 = A2*np.cos(w2*x + th2)
s3 = A3*np.cos(w3*x + th3)
s4 = A4*np.cos(w4*x + th4)



x = np.linspace(0.0, N*T, N)
y = s1 + s2 + s3 + s4




# Converting for frequency domain
yf = fft(y)
xf = np.linspace(0.0, 1.0/(2.0*T), N//2)


import matplotlib.pyplot as plt
plt.plot(xf, 2.0/N * np.abs(yf[0:N//2]))
plt.grid()
plt.show()

plt.plot(x[0:160],y[0:160])
plt.grid()
plt.show()
```

```python
#Activation implementation

filter = np.ones(len(xf))
filter[0:400] = 0
filter[1000:5200] = 0
plt.plot(xf, filter)
plt.show()
```

```python
plt.plot(xf, filter*2.0/N * np.abs(yf[0:N//2]))
```
